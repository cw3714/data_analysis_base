```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes = True)
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("wine_quality.csv")
```

```python
df.head()
```

```python
sns.set_theme(font_scale = 1.1, style = "ticks")
sns.pairplot(df, diag_kind = "kde")
plt.show()
```

```python
df.cov()
```

```python
df.corr(method = "pearson")
```

```python
sns.heatmap(df.corr(), cmap = "viridis")
```

```python
sns.clustermap(df.corr(), annot = True, cmap = "RdYlBu_r", vmin = -1, vmax = 1)
```

```python
np.triu(np.ones_like(df.corr()))
mask = np.triu(np.ones_like(df.corr(), dtype = np.bool_))
fig, ax = plt.subplots(figsize = (10, 10))
sns.heatmap(df.corr(), mask = mask, vmin = -1, vmax = 1, annot = True, cmap = "RdYlBu_r", cbar = True)
```
